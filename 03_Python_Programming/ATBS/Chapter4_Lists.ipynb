```python
## âœ… 1. Grocery List Maker â€“ with Functions
def get_items(prompt):
    items = []
    while True:
        item = input(prompt)
        if item == '':
            break
        items.append(item)
    return items

def print_list(items, title="Your list:"):
    print(f"\n{title}")
    for item in items:
        print("- " + item)

def main():
    grocery_list = get_items("Enter a grocery item (or press Enter to finish): ")
    print_list(grocery_list, "Your grocery list:")

main()
```

```python
## ðŸŽ¬ 2. Favorite Movies Collector â€“ with Functions

def collect_movies():
    movies = []
    
    while True:
        print("\nOptions: [add] movie, [remove] movie, [list] movies, [done] to finish")
        choice = input("What would you like to do?").strip().lower()

        if choice == "add":
            movie = input("Enter the movie name to add: ").strip()
            if movie:
                movies.append(movie)
                print(f"'{movie}' added")
        elif choice == "remove":
            movie = input("Enter the movie name to remove: ").strip()
            if movie in movies:
                movies.remove(movie)
                print(f"'{movie}' removed")
            else:
                print(f"'{movie} is not in your list.")
        elif choice == "list":
            display_movies(movies)
        elif choice == "done":
            break
        else:
            print("Invalid option. Try again.")
    
    return movies
        
def display_movies(movies):
    if not movies:
        print("Your movie list is empty.")
        return
        
    print("\nYour favorite movies are:")
    for movie in movies:
        print("* ", movie)


def main():
    movie_list = collect_movies()
    print("\nFinal movie list:")
    display_movies(movie_list)

main()
```

```python
## ðŸ§® 3. Average of Numbers â€“ with Functions
def get_numbers():
    numbers = []
    while True:
        entry = input("Enter a number (or press Enter to finish): ")
        if entry == '':
            break
        try:
            numbers.append(float(entry))
        except valueError:
            print("Invalid input. Please enter a valid number.")
    return numbers


def calculate_average(numbers):
    if numbers:
        return sum(numbers) / len(numbers)
    return None

def main():
    nums = get_numbers()
    avg = calculate_average(nums)
    if avg is not None:
        print(f"\nYou entered {len(nums)} numbers. The average is {avg:.2f}.")
    else:
        print("No valid numbers were entered.")

main()
```

```python
# How to Tell: Mutation vs. Reassignment
a = [1, 2, 3]
print(id(a))  # memory address

a.append(4)
print(id(a))  # same => mutation

a = [1, 2, 3, 4]
print(id(a))  # new => reassignment
```

```python
## 1. Reference vs. Copy
# Reference (no new object)
a = [1, 2, 3]
b = a  # b refers to the same list as a. Any change to a or b affects both. They share the same memory location (object).

# Copy (new object)
b = a.copy()      # shallow copy. Creates a new outer list. Shallow copy still shares inner objects.
c = copy.deepcopy(a)  # deep copy. Deep copy creates fully independent copies, even for inner structures.
```

```python
## 2. Mutation vs. Reassignment
# Mutation â†’ changes the object in place
a.append(4)          # Mutation
a[0] = 100           # Mutation
a[1].append(5)       # Mutation (if a is nested)
# Use methods like .append(), .sort(), .remove(), or indexing to change values.
# The id(a) (memory address) stays the same.

# Reassignment â†’ changes the reference
a = [4, 5, 6]        # Reassignment
a[0] = [9, 9]        # Reassigning an inner element
# It points the variable to a new object.
# Doesn't affect other references to the old list.
# id(a) will change after reassignment.
```

```python
# 3. Shallow vs. Deep Copy
# Shallow copy
a = [[1, 2], [3, 4]]
b = a.copy()
# a and b are different outer lists.
# But they share inner lists.
# Mutation of a[0][0] will affect b.
a[0][0] = 100  # both a and b are affected
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set parameters for the automaton
num_cells = 101  # Number of cells in the automaton
iterations = 51  # Number of iterations (generations)
initial_state = np.zeros(num_cells)  # Create an array of zeros
initial_state[num_cells // 2] = 1  # Set the middle cell to "1" (on)

# Function to apply Rule 30 to generate the next state
def apply_rule_30(state):
    new_state = np.zeros_like(state)  # Initialize an array of zeros (next generation)
    for i in range(1, len(state) - 1):  # Skip the first and last cells to avoid out-of-bounds
        left, center, right = state[i - 1], state[i], state[i + 1]
        # Apply Rule 30: center gets the new value based on neighbors
        new_state[i] = left ^ (center or right)  # XOR rule
    return new_state

# Set up an array to store all generations
generations = np.zeros((iterations, num_cells), dtype=int)
generations[0] = initial_state  # Set the first generation to the initial state

# Generate the automaton over time
for i in range(1, iterations):
    generations[i] = apply_rule_30(generations[i - 1])

# Plot the result using matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(generations, cmap='binary', interpolation='nearest')
plt.title("1D Cellular Automaton (Rule 30)")
plt.xlabel("Cell Index")
plt.ylabel("Generation")
plt.show()
```

```python
import random, time, copy

WIDTH = 5
HEIGHT = 5
NUM_ITERATIONS = 3  # Set the number of iterations (generations) you want to run.

# Create a list of list for the cells:
nextCells = []
for x in range(WIDTH):
    column = []  # Create a new column.
    for y in range(HEIGHT):
        if random.randint(0, 1) == 0:
            column.append('#')  # Add a living cell.
        else:
            column.append(' ')  # Add a dead cell.
    nextCells.append(column)  # nextCells is a list of column lists.

# Main loop, now limited to NUM_ITERATIONS iterations.
for iteration in range(NUM_ITERATIONS):
    print('\n\n\n\n\n')  # Separate each step with newlines.
    currentCells = copy.deepcopy(nextCells)

    # Print currentCells on the screen:
    for y in range(HEIGHT):
        for x in range(WIDTH):
            print(currentCells[x][y], end='')  # Print the # or space.
        print()  # Print a newline at the end of the row.

    # Calculate the next step's cells based on current step's cells:
    for x in range(WIDTH):
        for y in range(HEIGHT):
            # Get neighboring coordinates:
            # `% WIDTH` ensures leftCoord is always between 0 and WIDTH - 1
            leftCoord  = (x - 1) % WIDTH
            rightCoord = (x + 1) % WIDTH
            aboveCoord = (y - 1) % HEIGHT
            belowCoord = (y + 1) % HEIGHT

            # Count number of living neighbors:
            numNeighbors = 0
            if currentCells[leftCoord][aboveCoord] == '#':
                numNeighbors += 1 # Top-left neighbor is alive.
            if currentCells[x][aboveCoord] == '#':
                numNeighbors += 1 # Top neighbor is alive.
            if currentCells[rightCoord][aboveCoord] == '#':
                numNeighbors += 1 # Top-right neighbor is alive.
            if currentCells[leftCoord][y] == '#':
                numNeighbors += 1 # Left neighbor is alive.
            if currentCells[rightCoord][y] == '#':
                numNeighbors += 1 # Right neighbor is alive.
            if currentCells[leftCoord][belowCoord] == '#':
                numNeighbors += 1 # Bottom-left neighbor is alive.
            if currentCells[x][belowCoord] == '#':
                numNeighbors += 1 # Bottom neighbor is alive.
            if currentCells[rightCoord][belowCoord] == '#':
                numNeighbors += 1 # Bottom-right neighbor is alive.

            # Set cell based on Conway's Game of Life rules:
            if currentCells[x][y] == '#' and (numNeighbors == 2 or numNeighbors == 3):
                # Living cells with 2 or 3 neighbors stay alive:
                nextCells[x][y] = '#'
            elif currentCells[x][y] == ' ' and numNeighbors == 3:
                # Dead cells with 3 neighbors become alive:
                nextCells[x][y] = '#'
            else:
                # Everything else dies or stays dead:
                nextCells[x][y] = ' '

    time.sleep(1)  # Add a 1-second pause to reduce flickering.
```

```python
# What is the difference between copy.copy() and copy.deepcopy()?
# copy.copy() (Shallow Copy):
# it creates a new object, but it does not recursively copy the nested objects. Instead, it copies references to the nested objects.
# Effect: The top-level object is copied, but if the object contains other objects (e.g., lists inside lists, dictionaries inside lists, etc.)
# , the references to these nested objects are copied. Thus, changes to nested objects in the copied object will affect the original object 
# and vice versa.
import copy

original = [1, [2, 3]]
shallow_copy = copy.copy(original)

shallow_copy[0] = 99  # Modifying the top-level element
shallow_copy[1][0] = 88  # Modifying a nested element

print(original)        # Output: [1, [88, 3]]
print(shallow_copy)    # Output: [99, [88, 3]]
```

```python
def list_to_string(input_list):
    # Check if the list is empty
    if not input_list:
        return ""
    
    # If the list has only one item, return it
    if len(input_list) == 1:
        return input_list[0]
    
    # If the list has two items, just join them with 'and'
    if len(input_list) == 2:
        return " and ".join(input_list)
    
    # For longer lists, join items with commas and then insert 'and' before the last item
    return ", ".join(input_list[:-1]) + ", and " + input_list[-1]

# Testing the function
spam = ['apples', 'bananas', 'tofu', 'cats']
print(list_to_string(spam))  # Output: 'apples, bananas, tofu, and cats'

# Test with an empty list
empty_list = []
print(list_to_string(empty_list))  # Output: ''
```

```python
import random

# Initialize the counter for streaks
numberOfStreaks = 0

# Run the experiment 10,000 times
for experimentNumber in range(10000):
    # Generate a list of 100 random 'heads' or 'tails'
    coinFlips = [random.choice(['H', 'T']) for _ in range(100)]
    
    # Check if there is a streak of 6 heads or 6 tails in a row
    for i in range(len(coinFlips) - 5):  # len(coinFlips) - 5 ensures we can check 6 consecutive elements
        streak = coinFlips[i:i+6]
        if streak == ['H', 'H', 'H', 'H', 'H', 'H'] or streak == ['T', 'T', 'T', 'T', 'T', 'T']:
            numberOfStreaks += 1
            break  # Once we find a streak, we stop checking further in this experiment

# Calculate the percentage of experiments with a streak
print('Chance of streak: %s%%' % (numberOfStreaks / 100))
```
